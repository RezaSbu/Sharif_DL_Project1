```python
import cv2
import numpy as np
import tensorflow as tf

# مدل MobileNetV2 از ImageNet
model = tf.keras.applications.MobileNetV2(weights='imagenet')

preprocess_input = tf.keras.applications.mobilenet_v2.preprocess_input

decode_predictions = tf.keras.applications.mobilenet_v2.decode_predictions
```

```python
def predict_and_draw(frame):
    # تغییر اندازه به 224x224 برای MobileNetV2
    img = cv2.resize(frame, (224, 224))
    img_array = np.expand_dims(img, axis=0)
    img_array = preprocess_input(img_array)

    preds = model.predict(img_array)
    decoded = decode_predictions(preds, top=1)[0][0]  # (class_id, name, score)
    label = f"{decoded[1]}: {decoded[2]*100:.2f}%"

    cv2.putText(frame, label, (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    return frame
```

```python
cap = cv2.VideoCapture(0)  

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.flip(frame, 1)  
    frame = predict_and_draw(frame)

    cv2.imshow('Object Recognition', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):  
        break

cap.release()
cv2.destroyAllWindows()
```
